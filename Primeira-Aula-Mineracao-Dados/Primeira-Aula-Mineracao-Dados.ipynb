## Mineração de Dados

Mineração de dados é o processo de encontrar padrões, tendências e relações significativas ou interessantes em conjuntos de dados grandes e complexos. Envolve a aplicação de técnicas estatísticas, algoritmos de aprendizado de máquina e métodos de inteligência artificial para analisar grandes volumes de dados e extrair informações úteis e acionáveis.

## **Machine Learning**

É uma metodologia para análise de dados baseada na automatização construtiva de modelos analíticos. Na prática de usar algoritmos para coletar dados, aprender com eles, e então fazer uma determinação ou predição sobre alguma coisa no mundo. 

## **O que é aprendizado?**
É um processo que visa a capacidade de se adaptar, modificar e melhorar seu comportamento e respostas.

## **Por que processo?**
Porque é contínuo e evolutivo.

## **Como ocorre?**
Através de relação com o ambiente.

## **Como o computador aprende?**
 *   Interagindo com o ambiente, através de dados;
 *   Dados são usados para a máquina aprender;
 *   O aprendizado pode ser persistido através de um modelo;
 *   Contínuo, ou seja, o aprendizado deve melhorar. Com dados melhores, o computador pode melhorar seu modelo, ou seja, aprender mais sobre as condições do tempo
 *   O aprendizado pode ser medido. Podemos medir o quanto o computador aprendeu, por exemplo, em 100 previsões, acertou 83 e errou 17.

## **Tipos de aprendizagem**? 
*   **Aprendizado supervisionado : o sistema recebe um conjunto de dados em que a resposta correta já é prevista;**

  Exemplo: Classificacar se um aluno vai abandonar ou não um curso.
  
  
*   **Aprendizado não supervisionado: método é utilizado para dados que não possuem histórico. Ou seja, o sistema não possui uma resposta certa; cabe ao algoritmo interpretar o comando e explorar os dados para fornecer um resultado satisfatório;**

  Exemplo: Um exemplo de uso é nas pesquisas sobre hábitos de consumo de clientes, em que o sistema deve encontrar padrões ao cruzar diferentes variáveis, como registro de compras, frequência e perfil de cada cliente.
  
  
*   **Aprendizado semi-supervisionado:  modelo combina aspectos dos dois tipos de aprendizado que vimos até agora: supervisionado e não supervisionado;**

 Exemplo: Um exemplo de aplicação é o reconhecimento do rosto de uma pessoa em uma webcam.
 
 
*   **Aprendizado por reforço: o sistema não possui dados prévios; ele aprende por tentativa e erro, identificando quais ações geram as maiores recompensas. Cabe ao sistema coletar informações e se adaptar ao ambiente;**

  Exemplo: Como exemplo podemos tomar os softwares financeiros que montam portfólio de ações para investimentos.
  

*   **Aprendizagem por Transferência (Transfer Learning): Na aprendizagem por transferência, o conhecimento adquirido em uma tarefa é transferido e reaproveitado para melhorar o desempenho em uma tarefa relacionada. Isso é útil quando há poucos dados disponíveis para a nova tarefa ou quando as tarefas têm características semelhantes.**
